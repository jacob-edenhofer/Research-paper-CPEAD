```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skewnorm

# Define the grid for public opinion (X-axis) and cost-benefit ratio (Y-axis)
x = np.linspace(0, 10, 500)  # public opinion range from 0 to 10
y = np.linspace(0, 2, 500)   # cost-benefit ratio from 0 to 2, with 1 in the middle

# Function for generating a skewed normal distribution
def skewed_distribution(x, a, loc, scale):
    """a: skewness, loc: mean, scale: std dev"""
    return skewnorm.pdf(x, a, loc, scale)

# Scenario Definitions (Unique Variable Names)
# Scenario 1: Pro-Climate, Low Uncertainty, Beneficial (CB â‰¤ 1) (Green)
Z1_x_green = skewed_distribution(x, a=-2, loc=6, scale=0.5)
Z1_y_green = skewed_distribution(y, a=-3, loc=0.8, scale=0.2)

# Scenario 2: Pro-Climate, Low Uncertainty, Detrimental (CB > 1) (Green)
Z2_x_green = skewed_distribution(x, a=-1, loc=6.5, scale=0.7)
Z2_y_green = skewed_distribution(y, a=2, loc=1.2, scale=0.3)

# Scenario 3: Anti-Climate, Low Uncertainty, Beneficial (CB â‰¤ 1) (Green)
Z3_x_green = skewed_distribution(x, a=1, loc=3, scale=0.5)
Z3_y_green = skewed_distribution(y, a=-2, loc=0.9, scale=0.2)

# Scenario 4: Pro-Climate, High Uncertainty, Beneficial (CB â‰¤ 1) (Grey)
Z4_x_grey = skewed_distribution(x, a=-1, loc=7.5, scale=1.0)
Z4_y_grey = skewed_distribution(y, a=-1, loc=0.9, scale=0.5)

# Scenario 5: Anti-Climate, Low Uncertainty, Detrimental (CB > 1) (Grey)
Z5_x_grey = skewed_distribution(x, a=1, loc=2.5, scale=0.7)
Z5_y_grey = skewed_distribution(y, a=2, loc=1.3, scale=0.3)

# Scenario 6: Anti-Climate, High Uncertainty, Beneficial (CB â‰¤ 1) (Grey)
Z6_x_grey = skewed_distribution(x, a=2, loc=3, scale=1.0)
Z6_y_grey = skewed_distribution(y, a=-2, loc=0.8, scale=0.5)

# Function to plot scenarios with titles reflecting uncertainty
def plot_scenario_with_full_title(ax, Z_x, Z_y, title, median_x):
    Z_joint = np.outer(Z_y, Z_x)
    sns.heatmap(Z_joint, cmap="Blues", ax=ax, cbar=True, xticklabels=False, yticklabels=False)
    ax.set_title(title)
    ax.axvline(x=250, color='black', linestyle='--', label="CB Ratio = 1")
    ax.axvline(x=median_x, color='red', linestyle='--', label="Public Opinion Median")
    ax.axhline(y=250, color='black', linestyle='--')
    ax.set_xlabel("Public Opinion (Anti to Pro Climate)")
    ax.set_ylabel("Cost-Benefit Ratio (CB â‰¤ 1 beneficial, CB > 1 detrimental)")
    ax.legend()

# Plotting the six scenarios (two rows)
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
plot_scenario_with_full_title(axes[0, 0], Z1_x_green, Z1_y_green, "Pro-Climate, Low Uncertainty\n(Public Opinion), Low Uncertainty (CB â‰¤ 1)", median_x=300)
plot_scenario_with_full_title(axes[0, 1], Z2_x_green, Z2_y_green, "Pro-Climate, Low Uncertainty\n(Public Opinion), Low Uncertainty (CB > 1)", median_x=325)
plot_scenario_with_full_title(axes[0, 2], Z3_x_green, Z3_y_green, "Anti-Climate, Low Uncertainty\n(Public Opinion), Low Uncertainty (CB â‰¤ 1)", median_x=125)
plot_scenario_with_full_title(axes[1, 0], Z4_x_grey, Z4_y_grey, "Pro-Climate, High Uncertainty\n(Public Opinion), High Uncertainty (CB â‰¤ 1)", median_x=375)
plot_scenario_with_full_title(axes[1, 1], Z5_x_grey, Z5_y_grey, "Anti-Climate, Low Uncertainty\n(Public Opinion), Low Uncertainty (CB > 1)", median_x=100)
plot_scenario_with_full_title(axes[1, 2], Z6_x_grey, Z6_y_grey, "Anti-Climate, High Uncertainty\n(Public Opinion), High Uncertainty (CB â‰¤ 1)", median_x=150)

plt.tight_layout()
plt.savefig('green_grey_scenarios.png', figsize = c(9, 6))
plt.show()

# High Uncertainty Pro- and Anti-Climate Public Opinion (Unique Variable Names)
# Scenario 1: Pro-Climate, High Uncertainty, High Uncertainty (CB > 1)
Z7_x_high_unc = skewed_distribution(x, a=-1, loc=7, scale=1.0)
Z7_y_high_unc = skewed_distribution(y, a=2, loc=1.2, scale=0.5)

# Scenario 2: Pro-Climate, High Uncertainty, High Uncertainty (CB â‰¤ 1)
Z8_x_high_unc = skewed_distribution(x, a=-1, loc=7, scale=1.0)
Z8_y_high_unc = skewed_distribution(y, a=-2, loc=0.9, scale=0.5)

# Scenario 3: Anti-Climate, High Uncertainty, High Uncertainty (CB > 1)
Z9_x_high_unc = skewed_distribution(x, a=2, loc=3, scale=1.0)
Z9_y_high_unc = skewed_distribution(y, a=2, loc=1.3, scale=0.5)

# Scenario 4: Anti-Climate, High Uncertainty, High Uncertainty (CB â‰¤ 1)
Z10_x_high_unc = skewed_distribution(x, a=2, loc=3, scale=1.0)
Z10_y_high_unc = skewed_distribution(y, a=-2, loc=0.8, scale=0.5)

# Plot the four scenarios in a 2x2 grid
fig, axes_high_unc = plt.subplots(2, 2, figsize=(12, 10))
plot_scenario_with_full_title(axes_high_unc[0, 0], Z7_x_high_unc, Z7_y_high_unc, "Pro-Climate, High Uncertainty\n(Public Opinion), High Uncertainty (CB > 1)", median_x=350)
plot_scenario_with_full_title(axes_high_unc[0, 1], Z8_x_high_unc, Z8_y_high_unc, "Pro-Climate, High Uncertainty\n(Public Opinion), High Uncertainty (CB â‰¤ 1)", median_x=350)
plot_scenario_with_full_title(axes_high_unc[1, 0], Z9_x_high_unc, Z9_y_high_unc, "Anti-Climate, High Uncertainty\n(Public Opinion), High Uncertainty (CB > 1)", median_x=150)
plot_scenario_with_full_title(axes_high_unc[1, 1], Z10_x_high_unc, Z10_y_high_unc, "Anti-Climate, High Uncertainty\n(Public Opinion), High Uncertainty (CB â‰¤ 1)", median_x=150)

plt.tight_layout()
plt.savefig('high_uncertainty_red_scenarios.png')
plt.show()

# Purple Scenarios (Unique Variable Names)
# Scenario 1: Pro-Climate, Low Uncertainty (Public Opinion), Low Uncertainty (CB > 1)
Z11_x_purple = skewed_distribution(x, a=-1, loc=6.5, scale=0.5)
Z11_y_purple = skewed_distribution(y, a=2, loc=1.2, scale=0.2)

# Scenario 2: Anti-Climate, Low Uncertainty (Public Opinion), Low Uncertainty (CB â‰¤ 1)
Z12_x_purple = skewed_distribution(x, a=1, loc=3, scale=0.5)
Z12_y_purple = skewed_distribution(y, a=-2, loc=0.9, scale=0.2)

# Plot the two purple cell scenarios
fig, axes_purple = plt.subplots(1, 2, figsize=(12, 6))
plot_scenario_with_full_title(axes_purple[0], Z11_x_purple, Z11_y_purple, "Pro-Climate, Low Uncertainty\n(Public Opinion), Low Uncertainty (CB > 1)", median_x=325)
plot_scenario_with_full_title(axes_purple[1], Z12_x_purple, Z12_y_purple, "Anti-Climate, Low Uncertainty\n(Public Opinion), Low Uncertainty (CB â‰¤ 1)", median_x=175)

plt.tight_layout()
plt.savefig('purple_scenarios.png')
plt.show()

# Salmon and Blue Scenarios (Unique Variable Names)

# Salmon-Colored Cells
# Scenario 1: Anti-Climate, High Uncertainty (Public Opinion), Low Uncertainty (CB â‰¤ 1) (Beneficial)
Z13_x_salmon = skewed_distribution(x, a=2, loc=3, scale=1.0)  # Anti-climate with high uncertainty
Z13_y_salmon = skewed_distribution(y, a=-2, loc=0.9, scale=0.2)  # CB â‰¤ 1, low uncertainty

# Scenario 2: Pro-Climate, High Uncertainty (Public Opinion), Low Uncertainty (CB > 1) (Detrimental)
Z14_x_salmon = skewed_distribution(x, a=-1, loc=7, scale=1.0)  # Pro-climate with high uncertainty
Z14_y_salmon = skewed_distribution(y, a=2, loc=1.2, scale=0.2)  # CB > 1, low uncertainty

# Blue Cells
# Scenario 3: Anti-Climate, Low Uncertainty (Public Opinion), High Uncertainty (CB â‰¤ 1) (Beneficial)
Z15_x_blue = skewed_distribution(x, a=1, loc=3, scale=0.5)  # Anti-climate with low uncertainty
Z15_y_blue = skewed_distribution(y, a=-2, loc=0.9, scale=0.5)  # CB â‰¤ 1, high uncertainty

# Scenario 4: Pro-Climate, Low Uncertainty (Public Opinion), High Uncertainty (CB > 1) (Detrimental)
Z16_x_blue = skewed_distribution(x, a=-1, loc=7, scale=0.5)  # Pro-climate with low uncertainty
Z16_y_blue = skewed_distribution(y, a=2, loc=1.2, scale=0.5)  # CB > 1, high uncertainty

# Plotting function for salmon and blue cells
def plot_scenario_salmon_blue(ax, Z_x, Z_y, title, median_x, color):
    Z_joint = np.outer(Z_y, Z_x)
    sns.heatmap(Z_joint, cmap=color, ax=ax, cbar=True, xticklabels=False, yticklabels=False)
    ax.set_title(title)
    ax.axvline(x=250, color='black', linestyle='--', label="CB Ratio = 1")
    ax.axvline(x=median_x, color='red', linestyle='--', label="Public Opinion Median")
    ax.axhline(y=250, color='black', linestyle='--')
    ax.set_xlabel("Public Opinion (Anti to Pro Climate)")
    ax.set_ylabel("Cost-Benefit Ratio (CB â‰¤ 1 beneficial, CB > 1 detrimental)")
    ax.legend()

# Plotting the four scenarios (Salmon + Blue cells)
fig, axes_salmon_blue = plt.subplots(2, 2, figsize=(14, 12))

# Plot the two salmon-colored scenarios
plot_scenario_salmon_blue(axes_salmon_blue[0, 0], Z13_x_salmon, Z13_y_salmon, "Anti-Climate, High Uncertainty\n(Public Opinion), Low Uncertainty (CB â‰¤ 1)", median_x=150, color="Blues")
plot_scenario_salmon_blue(axes_salmon_blue[0, 1], Z14_x_salmon, Z14_y_salmon, "Pro-Climate, High Uncertainty\n(Public Opinion), Low Uncertainty (CB > 1)", median_x=350, color="Blues")

# Plot the two blue-colored scenarios
plot_scenario_salmon_blue(axes_salmon_blue[1, 0], Z15_x_blue, Z15_y_blue, "Anti-Climate, Low Uncertainty\n(Public Opinion), High Uncertainty (CB â‰¤ 1)", median_x=150, color="Blues")
plot_scenario_salmon_blue(axes_salmon_blue[1, 1], Z16_x_blue, Z16_y_blue, "Pro-Climate, Low Uncertainty\n(Public Opinion), High Uncertainty (CB > 1)", median_x=350, color="Blues")

plt.tight_layout()
# Save the last plot
plt.savefig('salmon_blue_scenarios.png')
plt.show()
```
